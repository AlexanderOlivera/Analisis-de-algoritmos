# 6. Validador del acertijo de Einstein

No modifiques esta sección.


```python
NACIONALIDADES = {"Británico", "Sueco", "Danés", "Noruego", "Alemán"}
COLORES = {"Roja", "Verde", "Blanca", "Amarilla", "Azul"}
BEBIDAS = {"Té", "Café", "Leche", "Cerveza", "Agua"}
CIGARROS = {"Pall Mall", "Dunhill", "Blend", "BlueMaster", "Prince"}
MASCOTAS = {"Perros", "Pájaros", "Gatos", "Caballos", "Peces"}

def validar_formato(casas):
    if not isinstance(casas, list) or len(casas) != 5:
        return False, "La solución debe ser una lista de 5 casas."

    required = {"posicion", "nacionalidad", "color", "bebida", "cigarro", "mascota"}

    for casa in casas:
        if not isinstance(casa, dict):
            return False, "Cada casa debe representarse como un diccionario."
        if set(casa.keys()) != required:
            return False, f"Cada casa debe tener exactamente las llaves: {required}"

    posiciones = [c["posicion"] for c in casas]
    if set(posiciones) != {1, 2, 3, 4, 5}:
        return False, "Las posiciones deben ser 1, 2, 3, 4 y 5."

    casas = sorted(casas, key=lambda x: x["posicion"])

    if set(c["nacionalidad"] for c in casas) != NACIONALIDADES:
        return False, "Las nacionalidades no coinciden o hay repetidas."
    if set(c["color"] for c in casas) != COLORES:
        return False, "Los colores no coinciden o hay repetidos."
    if set(c["bebida"] for c in casas) != BEBIDAS:
        return False, "Las bebidas no coinciden o hay repetidas."
    if set(c["cigarro"] for c in casas) != CIGARROS:
        return False, "Los cigarros no coinciden o hay repetidos."
    if set(c["mascota"] for c in casas) != MASCOTAS:
        return False, "Las mascotas no coinciden o hay repetidas."

    return True, "Formato correcto."

def find_house(casas, atributo, valor):
    for casa in casas:
        if casa[atributo] == valor:
            return casa
    return None

def same_house(casas, atributo1, valor1, atributo2, valor2):
    c1 = find_house(casas, atributo1, valor1)
    c2 = find_house(casas, atributo2, valor2)
    return c1["posicion"] == c2["posicion"]

def next_to(casas, atributo1, valor1, atributo2, valor2):
    c1 = find_house(casas, atributo1, valor1)
    c2 = find_house(casas, atributo2, valor2)
    return abs(c1["posicion"] - c2["posicion"]) == 1

def immediately_left_of(casas, atributo1, valor1, atributo2, valor2):
    c1 = find_house(casas, atributo1, valor1)
    c2 = find_house(casas, atributo2, valor2)
    return c1["posicion"] + 1 == c2["posicion"]

def validar_acertijo_einstein(casas, verbose=True):
    ok, msg = validar_formato(casas)
    if not ok:
        if verbose:
            print("❌", msg)
        return False

    casas = sorted(casas, key=lambda x: x["posicion"])
    checks = []

    checks.append(("El británico vive en la casa roja", same_house(casas, "nacionalidad", "Británico", "color", "Roja")))
    checks.append(("El sueco tiene perros", same_house(casas, "nacionalidad", "Sueco", "mascota", "Perros")))
    checks.append(("El danés bebe té", same_house(casas, "nacionalidad", "Danés", "bebida", "Té")))
    checks.append(("La casa verde está inmediatamente a la izquierda de la blanca", immediately_left_of(casas, "color", "Verde", "color", "Blanca")))
    checks.append(("El dueño de la casa verde bebe café", same_house(casas, "color", "Verde", "bebida", "Café")))
    checks.append(("Quien fuma Pall Mall tiene pájaros", same_house(casas, "cigarro", "Pall Mall", "mascota", "Pájaros")))
    checks.append(("El dueño de la casa amarilla fuma Dunhill", same_house(casas, "color", "Amarilla", "cigarro", "Dunhill")))
    checks.append(("El hombre del centro bebe leche", find_house(casas, "bebida", "Leche")["posicion"] == 3))
    checks.append(("El noruego vive en la primera casa", find_house(casas, "nacionalidad", "Noruego")["posicion"] == 1))
    checks.append(("Quien fuma Blend vive junto al que tiene gatos", next_to(casas, "cigarro", "Blend", "mascota", "Gatos")))
    checks.append(("Quien tiene caballos vive junto al que fuma Dunhill", next_to(casas, "mascota", "Caballos", "cigarro", "Dunhill")))
    checks.append(("Quien fuma BlueMaster bebe cerveza", same_house(casas, "cigarro", "BlueMaster", "bebida", "Cerveza")))
    checks.append(("El alemán fuma Prince", same_house(casas, "nacionalidad", "Alemán", "cigarro", "Prince")))
    checks.append(("El noruego vive junto a la casa azul", next_to(casas, "nacionalidad", "Noruego", "color", "Azul")))
    checks.append(("Quien fuma Blend tiene un vecino que bebe agua", next_to(casas, "cigarro", "Blend", "bebida", "Agua")))

    total_ok = True
    for texto, resultado in checks:
        if verbose:
            print(("✅" if resultado else "❌"), texto)
        total_ok = total_ok and resultado

    if verbose:
        if total_ok:
            pez = find_house(casas, "mascota", "Peces")
            print("\n🎉 Todas las pistas se cumplen.")
            print("La persona que tiene los peces es:", pez["nacionalidad"])
        else:
            print("\nLa solución no cumple todas las restricciones.")

    return total_ok
```
